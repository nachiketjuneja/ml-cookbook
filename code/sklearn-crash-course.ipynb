```python
# load iris dataset (flower classification)
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

x,y=load_iris(return_X_y=True)
y
```

```python
## There could be possibilites where you dump entire dataset into training data and the model ends up memorizing it instead of being traines
## When actual scenarios come, the model might not behave as expected since it memorized stuff due to the overfitting with training data
## So we train on one portion of data and evaluate on the other portion so that it doesn't memorize.
## can use validation dataset also sometimes

## Perform a random split on the dataset
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) # 20% test, 80% train
```

```python
len(x_train)
len(x_test)
```

```python
## The problem with the above is that it is a random split. What if the test data had very limited of a specific class, or the distribution of data between
## test and train varied heavily, or some class wasn't there at all.
## to minimize this, we can use below to guarentee similar distribution. Will have similar ratio.

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

for train_idx, test_idx in split.split(x,y):
  x_train, x_test = x[train_idx], x[test_idx]
  y_train, y_test = y[train_idx], y[test_idx]

# There are other split options too, refer documentation
```

```python
## PRE PROCESSING
```

```python
## What if there's preprocessing required on data before it can be used for training or testing

## scale data to a specific format. For ex, with KNN, we need to scale them to same units / variants
## StandardScaler = substracts mean from the data and divides by standard deviation
from sklearn.preprocessing import StandardScaler

#before
x,y = load_iris(return_X_y=True)
x
```

```python
#after
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train) # fit transform means learn from train dataset, and scale the values as per training data
x_test_scaled = scaler.transform(x_test) # only transform in this case since we already learnt and we want to use the same mean / variance in this case

x_train_scaled
```

```python
## Use numpy as evidence how scaling worked

import numpy as np

(x_train - np.mean(x_train, axis=0))/np.std(x_train, axis=0) # axis=0 because we want this per feature
```

```python
## one hot encoder
## what if you wanted to have cols with value 0 or 1. For ex, occupation, you might create a col with multiple occupations
## and have the value as 0 or 1 if the person is in that occupation

from sklearn.preprocessing import OneHotEncoder

adult.occupation.value_counts()
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False) # sparse as false so that we get individual cols
encoded_vals = encoder.fit_transform(adult[['occupation', 'race']])
new_cols = encoder.get_feature_names_out(['occupation', 'race'])

new_cols # observe race_ cols or occupation_ cols
```

```python
df_encoded = pd.DataFrame(encoded_vals, columns=new_cols, index=adult.index)
adult_final = pd.concat(
    [adult.drop(columns=['occupation', 'race']), df_encoded], axis=1
)
adult_final
```

```python
## CLASSIFICATION EXAMPLE
## Code remains similar for various algorithms
## Depending on which alog we are using, we can tune the hyper parameters. This will be important for optimizations
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x,y = load_breast_cancer(return_X_y=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
scaler=StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# import classifier

from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier()
clf.fit(x_train_scaled, y_train) #train the model on the training data
```

```python
clf.score(x_test_scaled, y_test) # how well it performs on unseen data
```

```python
# Make predictions
single_instance = x_test_scaled[0]
single_instance

clf.predict([single_instance])
```

```python
# validate above answer from y_test
y_test[0]
```

```python
## Predict with probability

clf.predict_proba([single_instance]) # 0% 0, 100% 1
```

```python
## Logistic Regression

from sklearn.linear_model import LogisticRegression

log_clf = LogisticRegression()
log_clf.fit(x_train_scaled, y_train)
log_clf.score(x_test_scaled, y_test)
log_clf.predict([single_instance])
```

```python
## Decision Tree classifier

from sklearn.tree import DecisionTreeClassifier

dec_clf = DecisionTreeClassifier()
dec_clf.fit(x_train_scaled, y_train)
dec_clf.score(x_test_scaled, y_test)
dec_clf.predict([single_instance])
```

```python
## SVM

from sklearn.svm import SVC

svm_clf = SVC()
svm_clf.fit(x_train_scaled, y_train)
svm_clf.score(x_test_scaled, y_test)
svm_clf.predict([single_instance])
```

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier()
rf_clf.fit(x_train_scaled, y_train)
rf_clf.score(x_test_scaled, y_test)
rf_clf.predict([single_instance])
```

```python
## Naive Baise

from sklearn.naive_bayes import GaussianNB

gnb_clf = GaussianNB()
gnb_clf.fit(x_train_scaled, y_train)
gnb_clf.score(x_test_scaled, y_test)
gnb_clf.predict([single_instance])
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

x,y = fetch_california_housing(return_X_y=True)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
scaler=StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()
lreg.fit(x_train_scaled, y_train)
lreg.score(x_test_scaled, y_test)
lreg.predict([x_test_scaled[0]])
```

```python
## Lasso Ridge

from sklearn.linear_model import Lasso, Ridge
lasso = Lasso()
lasso.fit(x_train_scaled, y_train)
lasso.score(x_test_scaled, y_test)
lasso.predict([x_test_scaled[0]])

ridge = Ridge()
ridge.fit(x_train_scaled, y_train)
ridge.score(x_test_scaled, y_test)
ridge.predict([x_test_scaled[0]])
```
